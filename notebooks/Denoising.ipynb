```python
import os, cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
file_list = os.listdir("../src/images/")
latest_num = sorted(list(map(lambda f: int(f.split('.')[0]), file_list)))[-1]
latest_file = str(latest_num)+".png"
latest_file
```

```python
img = cv2.imread("../src/images/"+latest_file)
print(img.shape)
plt.imshow(img)
```

```python
kernel = np.ones((2, 2), np.uint8)
img = cv2.erode(img, kernel, iterations=1)
img = cv2.dilate(img, kernel, iterations=1)
plt.imshow(img)
```

```python
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
img = cv2.filter2D(img, -1, kernel)
plt.axis('off')
plt.imshow(img)
```
